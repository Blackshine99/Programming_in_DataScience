# Image Processing using Numpy and Matplotlib

## > Importing the necessary libraries

```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

## > Task 1: Read the Image from URL

```python
img_path = 'https://homepages.cae.wisc.edu/~ece533/images/airplane.png'
pic = plt.imread(img_path)
plt.imshow(pic)
plt.axis('off')
```

## > Separating out the Red Blue and Green channels and visualising them

```python
shape = pic.shape

#red channel
red_img = pic[:,:,0]

#green channel
green_img = pic[:,:,1]

#blue channel
blue_img = pic[:,:,2]
```

```python
#red channel
fig, ax = plt.subplots(1, 3, figsize=(30, 10))
plt.subplot(1, 3, 1)
temp1 = np.zeros(shape)
temp1[:, :, 0] = red_img
plt.imshow(temp1)
plt.title("red channel", fontdict = {'fontsize' : 50})

#green channel
plt.subplot(1, 3, 2)
temp2 = np.zeros(shape)
temp2[:, :, 1] = green_img
plt.imshow(temp2)
plt.title("green channel", fontdict = {'fontsize' : 50})

#blue channel
plt.subplot(1, 3, 3)
temp3 = np.zeros(shape)
temp3[:, :, 2] = blue_img
plt.imshow(temp3)
plt.title("blue channel", fontdict = {'fontsize' : 50})
```

## > Task 2: Find the Histogram of the Image and Plot it

```python
plt.hist(pic.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k')
plt.title('Histogram')
plt.show()
```

## > Task 3: Applying a mask to change color of image

This is shown by 2 methods: 
1. By using numpy array slicing to multiply each channel by a value
2. By using np.dot() to multiply the matrix by a (3,1) row vector

### - Method 1: 

```python
new_pic = pic.copy()
#multiplying each channel by a number to modify their values and change the color
new_pic[:,:, 0] = 0.2*pic[:,:,0]
new_pic[:,:, 1] = 0.1*pic[:,:,1]
new_pic[:,:, 2] = 0.4*pic[:,:,2]
```

```python
new_pic.shape
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(new_pic)
plt.title("Colored image", fontdict = {'fontsize' : 30})
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(pic)
plt.title("Original image", fontdict = {'fontsize' : 30})
plt.axis('off')
```

### - Method 2: 

```python
new_pic2 = np.dot(pic, [0.1, 0.4, 0.0])
fig, ax = plt.subplots(1, 2, figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(new_pic2)
plt.title("Colored image", fontdict = {'fontsize' : 30})
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(pic)
plt.title("Original image", fontdict = {'fontsize' : 30})
plt.axis('off')
```

```python
new_pic2.shape
```

### Task 4: Convert RGB image into GrayScale Image 
This can be done by 2 methods:
1. Taking average of all 3 channels
2. Taking weighted average of all 3 channels

### - Method 1:

```python

gray_pic1 = np.zeros((pic.shape[0], pic.shape[1]))
gray_pic1 = (red_img + blue_img + green_img)/3

fig, ax = plt.subplots(1, 2, figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(gray_pic1, cmap=plt.get_cmap("gray"))
plt.title("GrayScale image", fontdict = {'fontsize' : 30})
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(pic)
plt.title("Original image", fontdict = {'fontsize' : 30})
plt.axis('off')
```

### - Method 2: 

```python
gray_pic1 = np.zeros((pic.shape[0], pic.shape[1]))
gray_pic1 = 0.1*red_img + 0.4*blue_img + 0.5*green_img

fig, ax = plt.subplots(1, 2, figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(gray_pic1, cmap=plt.get_cmap("gray"))
plt.title("GrayScale image", fontdict = {'fontsize' : 30})
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(pic)
plt.title("Original image", fontdict = {'fontsize' : 30})
plt.axis('off')
```
